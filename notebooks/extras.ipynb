```python
import pandas as pd
import numpy as np
import seaborn as sns
import requests
from matplotlib import pyplot as plt
```

```python
df2 = pd.read_csv(r'C:\Users\yeray\notebooks\modificaciones sobremesa\Rampup\2-Data_Analysis\Entregas\EDA\data\peliculas.csv')
df1 = pd.read_csv(r'C:\Users\yeray\notebooks\modificaciones sobremesa\Rampup\2-Data_Analysis\Entregas\EDA\data\megadata.csv')
```

```python
df1['Generos'].fillna('Desconocido', inplace=True)
df1['Lanzamiento'] = pd.to_datetime(df1['Lanzamiento'])
df1['Idiomas hablados'].fillna('Desconocido', inplace=True)
df1 = df1[df1['Lanzamiento'].dt.year < 2020]
df1['Titulo'] = df1['Titulo'].astype(str)
df1 = df1[~df1['Titulo'].str.contains('sex', case=False)]
```

```python
peliculas_mas_populares = df2.nlargest(10, 'Popularidad')
sns.set_style('whitegrid')
plt.figure(figsize=(10, 5))
plt.bar(peliculas_mas_populares['Titulo'], peliculas_mas_populares['Popularidad'])
plt.title('Películas más populares')
plt.xlabel('Película')
plt.ylabel('Popularidad')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df1['Lanzamiento'], bins=50, color='cyan', alpha=0.8)
plt.axvline(df1['Lanzamiento'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.axvline(df1['Lanzamiento'].median(), color='green', linestyle='dashed', linewidth=1)
plt.title('Histograma de la fecha de lanzamiento')
plt.xlabel('Fecha de lanzamiento')
plt.ylabel('Cantidad')
plt.legend(['Media', 'Mediana'])
plt.show()
```

```python
sns.heatmap(df2.corr(), cmap='coolwarm', annot=True)
plt.title('Heatmap de todo el DataFrame')
plt.show()
```
