```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
import seaborn as sns
import requests
from matplotlib import pyplot as plt
```

```python
datospelis = []
```

```python
url = "https://api.themoviedb.org/3/movie/{}?api_key=7b34ecc56572df48f2ce676be26adbeb"

for movie_id in range(1, 500000):

    # Reemplaza el corchete {} en la URL con el valor actual de movie_id
    movie_url = url.format(movie_id)

    # Realiza la solicitud a la API
    response = requests.get(movie_url)

    # Si la respuesta es exitosa, imprime el título de la película
    if response.status_code == 200:
        data = response.json()
        datospelis.append({
            'ID': data['id'],
            'Titulo': data['title'],
            'Generos': ','.join(g['name'] for g in data['genres']),
            'Idioma original': data['original_language'],
            'Popularidad': data['popularity'],
            'Lanzamiento': data['release_date'],
            'Presupuesto': data['budget'],
            'Duracion': data['runtime'],
            'Sinopsis': data['overview'],
            'Puntuacion': data['vote_average'],
            'Numero de votos': data['vote_count'],
            'Idiomas hablados': ','.join([l['name'] for l in data['spoken_languages']]),
            'Imagen de portada': data['poster_path']
        })
    else:
        print("Error al obtener la información de la película con el ID:", movie_id)
```

```python
df = pd.DataFrame(datospelis)
```
